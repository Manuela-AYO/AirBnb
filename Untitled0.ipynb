```python
def hex_to_rgb(value):
    """
    Converts hex to rgb colours
    value: string of 6 characters representing a hex colour.
    Returns: list length 3 of RGB values
    """
    value = value.strip("#")  # removes hash symbol if present
    lv = len(value)
    return tuple(int(value[i:i + lv // 3], 16) for i in range(0, lv, lv // 3))


def rgb_to_dec(value):
    """
    Converts rgb to decimal colours (i.e. divides each value by 256)
    value: list (length 3) of RGB values
    Returns: list (length 3) of decimal values
    """
    return [v / 256 for v in value]


def get_continuous_cmap(hex_list, float_list=None):
    """
    Creates and returns a color map that can be used in heat map figures.
    If float_list is not provided, colour map graduates linearly between each color in hex_list.
    If float_list is provided, each color in hex_list is mapped to the respective location in float_list.

    Parameters
    ----------
    hex_list: list of hex code strings
    float_list: list of floats between 0 and 1, same length as hex_list. Must start with 0 and end with 1.

    Returns
    ----------
    colour map
    """
    rgb_list = [rgb_to_dec(hex_to_rgb(i)) for i in hex_list]
    if float_list:
        pass
    else:
        float_list = list(np.linspace(0, 1, len(rgb_list)))

    cdict = dict()
    for num, col in enumerate(['red', 'green', 'blue']):
        col_list = [[float_list[i], rgb_list[i][num], rgb_list[i][num]] for i in range(len(float_list))]
        cdict[col] = col_list
    cmp = mcolors.LinearSegmentedColormap('my_cmp', segmentdata=cdict, N=256)
    return cmp


def plot_colortable(hex_list):
    """
    Given a list of HEX strings, display the color as a cell, alongside the HEX and RGB values of the color.
    """
    cell_width = 150
    cell_height = 50
    swatch_width = 48
    margin = 0
    topmargin = 40

    rgb_list = [hex_to_rgb(value) for value in hex_list]
    dec_list = [rgb_to_dec(value) for value in rgb_list]
    names = [f'HEX: {col[0]}\nRGB: {col[1]}' for col in zip(hex_list, rgb_list, dec_list)]
    n = len(names)
    ncols = 4
    nrows = n // ncols + int(n % ncols > 0)

    width = cell_width * 8 + 2 * margin
    height = cell_height * nrows + margin + topmargin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin / width, margin / height,
                        (width - margin) / width, (height - topmargin) / height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows - 0.5), -cell_height / 2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        swatch_end_x = cell_width * col + swatch_width
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.hlines(y, swatch_start_x, swatch_end_x,
                  color=dec_list[i], linewidth=18)
    return fig

nkhl21_clrs = ['#040000ff', '#440000ff', '#600000ff', '#763000ff',
               '#935d00ff', '#cdb67bff', '#ebe3cbff', '#fdfffdff']
my_cmap = get_continuous_cmap(nkhl21_clrs)

# End custom colormap #

def plot_field_3D(field,  fig=None, ax=None,  aspect=(1, 1, 0.2, 1), save_path=None, cmap=my_cmap, colorbar = True):
    """
    The first three numbers of aspect are the scalings of x, y, z.
    """
    if ax:
       ax.get_proj = lambda: np.dot(Axes3D.get_proj(ax), np.diag(aspect))
    else:
        fig, ax = plt.subplots(subplot_kw={"projection": "3d", "proj_type": 'ortho'}, figsize=(10, 10))
        ax.get_proj = lambda: np.dot(Axes3D.get_proj(ax), np.diag(aspect))


    Y = np.arange(0, field.shape[1], 1)
    X = np.arange(0, field.shape[0], 1)

    X, Y = np.meshgrid(X, Y)
    Z = field

    # Plot the surface.

    surf = ax.plot_surface(X, Y, Z, cmap=cmap,
                           linewidth=0, antialiased=True,
                           rstride=1, cstride=1, edgecolor='none')

    # Customize the z axis.
    ax.set_zlim(field.min()*1.1, field.max()*1.1)
    ax.set_zticks([field.min(), 0, field.max()])
    plt.tick_params(
         axis='z',  # changes apply to the z-axis
         which='both',  # both major and minor ticks are affected
         bottom=False,  # ticks along the bottom edge are off
         top=False,  # ticks along the top edge are off
         labelbottom=False)

    ax.zaxis._axinfo["grid"].update({"linewidth":1})

    # Add a color bar which maps values to colors.
    if colorbar:
        fig.colorbar(surf, shrink=0.2, aspect=10)

    if save_path:
        plt.savefig(save_path, dpi=900)
    else:
        return ax
    plt.close(fig)


def plot_field_2D(field, fig=None, ax=None, save_path=None, cmap=my_cmap, colorbar=False):
    if ax:
        fig.patch.set_visible(False)
        ax.axis('off')
        pattern = ax.imshow(field, cmap=cmap)
        if colorbar:
            fig.colorbar(pattern, fraction=0.046, pad=0.04)
        return ax
    else:
        fig, ax = plt.subplots(figsize=(4, 4))
        fig.patch.set_visible(False)
        ax.axis('off')
        pattern = ax.imshow(field, cmap=cmap)
        if colorbar:
            fig.colorbar(pattern, fraction=0.046, pad=0.04)

    if save_path:
        plt.savefig(save_path, dpi=900)
    else:
        return ax

    plt.close(fig)


def plot_field_lyapunov(u, logscale=False, ax=None, fontsize=16, labelsize=16, linewidth=3):
    df = pd.DataFrame(columns=["Lyapunov functional", "Time"])
    df["Lyapunov functional"] = u.transient_lyaps
    df["Time"] = u.transient_times

    if logscale:
        df["Lyapunov functional"] += np.abs(np.min(df["Lyapunov functional"]))
        df.plot(x="Time", y="Lyapunov functional", logy=True, logx=True, zorder=1, ax=ax, legend=False,
                linewidth=linewidth)
    else:
        df.plot(x="Time", y="Lyapunov functional", zorder=1, ax=ax, legend=False, linewidth=linewidth)

    ax.tick_params(axis='both', which='major', labelsize=labelsize)
    ax.set_xlabel("Time", size=20)

    return ax


def plot_field_lyapunovs(u, savepath=None):
    fig, axs = plt.subplots(1, 2, figsize=(16, 5))
    plt.suptitle("Lyapunov functional time evolution", fontsize=24)
    plot_field_lyapunov(u, False, axs[0], fontsize=20, labelsize=20, linewidth=3)
    axs[0].set_ylabel("Lyap.", size=20)

    axs[1].set_ylabel("Lyap.+ min(Lyap.)", size=20)
    plot_field_lyapunov(u, True, axs[1], fontsize=20, labelsize=20, linewidth=3)
    plt.subplots_adjust(wspace=0.4)
    if savepath:
        plt.savefig(savepath, dpi=900)
    else:
        plt.show()
    plt.close(fig)
```

```python
class PinnSH:
    def __init__(self, xy0 : np.array, xy_db : np.array, xy_ub : np.array, xy_rb : np.array, xy_lb : np.array,  sol0 : np.array, XY_f : np.array, layers : list, epsilon : float, gamma : float) -> None:
        """
            xy0(np.array) : 
        """
        self.epsilon = epsilon # bifurcation parameter
        self.gamma = gamma  # pertubation parameter

        # initial points tensors
        self.x0 = torch.tensor(xy0[:, 0:1], dtype=dtype, requires_grad=True, device=device)
        self.y0 = torch.tensor(xy0[:, 1:2], dtype=dtype, requires_grad=True, device=device)
        self.t0 = torch.tensor(xy0[:, 2:3], dtype=dtype, requires_grad=True, device=device)
        
        # lower/down border tensors(x, 0, t)
        self.x_db = torch.tensor(xy_db[:, 0:1], dtype=dtype, requires_grad=True, device=device)
        self.y_db = torch.tensor(xy_db[:, 1:2], dtype=dtype, requires_grad=True, device=device)
        self.t_db = torch.tensor(xy_db[:, 2:3], dtype=dtype, requires_grad=True, device=device)

        # upper border tensors(x, len(y), t)
        self.x_ub = torch.tensor(xy_ub[:, 0:1], dtype=dtype, requires_grad=True, device=device)
        self.y_ub = torch.tensor(xy_ub[:, 1:2], dtype=dtype, requires_grad=True, device=device)
        self.t_ub = torch.tensor(xy_ub[:, 2:3], dtype=dtype, requires_grad=True, device=device)

        # right border tensors(0, len(y), t)
        self.x_rb = torch.tensor(xy_rb[:, 0:1], dtype=dtype, requires_grad=True, device=device)
        self.y_rb = torch.tensor(xy_rb[:, 1:2], dtype=dtype, requires_grad=True, device=device)
        self.t_rb = torch.tensor(xy_rb[:, 2:3], dtype=dtype, requires_grad=True, device=device)

        # left border tensors(x.shape, y, t)
        self.x_lb = torch.tensor(xy_lb[:, 0:1], dtype=dtype, requires_grad=True, device=device)
        self.y_lb = torch.tensor(xy_lb[:, 1:2], dtype=dtype, requires_grad=True, device=device)
        self.t_lb = torch.tensor(xy_lb[:, 2:3], dtype=dtype, requires_grad=True, device=device)

        # collocation points
        self.x_f = torch.tensor(XY_f[:, 0:1], dtype=dtype, requires_grad=True, device=device)
        self.y_f = torch.tensor(XY_f[:, 1:2], dtype=dtype, requires_grad=True, device=device)
        self.t_f = torch.tensor(XY_f[:, 2:3], dtype=dtype, requires_grad=True, device=device)

        # some field values at t=0
        self.sol0 = torch.tensor(sol0, dtype=dtype, device=device)

        # let us build the model
        self.model = self.build_model(layers)

        # to keep our train history
        self.train_cost_history = []


    def init_weights(self, m):
      if isinstance(m, nn.Linear):
          torch.nn.init.kaiming_uniform_(m.weight)
          m.bias.data.fill_(0.01)
  
    def build_model(self, layers : list) -> nn.Sequential :
        activation = nn.Tanh()
        network = []
        for i in range(len(layers)-2):
            network.append(
                nn.Linear(layers[i], layers[i+1])
                )
            network.append(activation)
        network.append(nn.Linear(layers[i+1], layers[i+2]))
        model = nn.Sequential(*network).to(device)
        model.apply(self.init_weights)
        return model
    

    def field_u(self, x, y, t):
        return self.model(torch.cat((x,y,t),1))


    def __compute_gradient(self, u, coord, n):
        if n == 0:
            return u
        du_dcoord = grad(u, coord, torch.ones_like(u).to(device),
                         retain_graph=True, create_graph=True)[0]
        return self.__compute_gradient(du_dcoord, coord, n-1)

    def pde(self, x, y, t):
        u = self.field_u(x, y, t) 

        # derivative with respect to t
        u_t = self.__compute_gradient(u, t, 1)

        # second order derivative with respect to x and y
        u_xx = self.__compute_gradient(u, x, 2)
        u_yy = self.__compute_gradient(u, y, 2)

        # fourth order derivative with respect to x and y
        u_xxxx = self.__compute_gradient(u, x, 4)
        u_yyyy = self.__compute_gradient(u, y, 4)

        # return the residual SH pde
        return u_t - self.epsilon*u + (u_xxxx + u_yyyy+ 2*(u_xx + u_yy) + u) - self.gamma*u**2 + u**3
    

    def __periodic_boundary(self):
      pass
    
    def loss_function(self) -> float:

        # prediction on initial conditions
        sol0_pred = self.field_u(self.x0, self.y0, self.t0)

        # prediction on boundaries
        """
        db_pred = self.field_u(self.x_db, self.y_db, self.t_db)
        ub_pred = self.field_u(self.x_ub, self.y_ub, self.t_ub)
        rb_pred = self.field_u(self.x_rb, self.y_rb, self.t_rb)
        lb_pred = self.field_u(self.x_lb, self.y_lb, self.t_lb)
        """

        # constraint on the pde
        colloc_pred = self.pde(self.x_f, self.y_f, self.t_f)

        # different losses
        loss_initial = torch.mean((sol0_pred - self.sol0)**2)
        loss_colloc = torch.mean((colloc_pred)**2)

        # compute the mean squared error on those predictions
        return loss_initial , loss_colloc # + \
            # torch.mean((db_pred - ub_pred)**2) + \
            # torch.mean((rb_pred - lb_pred)**2) + \
            # torch.mean((colloc_pred)**2)
    

    def train(self, epochs, **kwargs):
        optimizer = torch.optim.Adam(self.model.parameters(), **kwargs)

        for epoch in range(epochs):
            # compute the loss
            loss_initial, loss_colloc = self.loss_function()
            loss = loss_initial + loss_colloc
            # loss = self.loss_function()

            # save the loss at this time step
            self.train_cost_history.append([(loss_initial.cpu().detach(), loss_colloc.cpu().detach())])
            # self.train_cost_history.append([loss.cpu().detach()])

            # set the gradients to zero to avoid accumulation
            optimizer.zero_grad()

            # backprop
            loss.backward(retain_graph=True)
            
            # step
            optimizer.step()

            # quick display to see where we are
            if epoch % 100 == 0:
                print(f'Epoch ({optimizer}): {epoch}, Loss: {loss.detach().cpu().numpy()}')
```

```python
def make_test(n : int, t : float):
  x = y = np.arange(n)
  X,Y = np.meshgrid(x,y)
  xyt = np.hstack((X.flatten()[:,None], Y.flatten()[:,None], t*(np.ones(X.flatten()[:,None].shape))))
  return xyt
```

```python
if __name__ == "__main__":
    torch.cuda.empty_cache()

    np.random.seed(1)

    nb_initial = 50 # number of initial points
    nb_colloc = 20000 # number of collocation points

    # read the csv to get epsilon, gamma and times
    with open("/content/Data/data.csv") as file :
        lines = csv.DictReader(file)
        for row in lines :
            n = int(row["n"])
            epsilon = float(row["ε"])
            gamma = float(row["g"])
            times = row["times"]

    times = ast.literal_eval(times)

    # random indices for initial
    """
    idx_x = np.random.choice(n, nb_initial, replace=False)
    idx_y = np.random.choice(n, nb_initial, replace=False)
    xy = np.hstack((idx_x[:,None], idx_y[:,None]))
    """
    x,y = np.meshgrid(np.arange(n), np.arange(n))
    xy = np.hstack((x.flatten()[:,None], y.flatten()[:,None]))
    
    # (x,y,t)-coordinates at t=0
    xy0 = np.concatenate((xy, np.zeros((xy.shape[0],1))), 1)
    
    # values at thoses indices
    initial_array = np.load("/content/Data/initial.npy")
    """
    values = initial_array[idx_x, idx_y]
    values = values.reshape(-1, 1)
    """
    values = initial_array.reshape(-1,1)
    
    # first step : ok. now let's dive into those bounds

    # lower border points (x, 0, t)
    times = [0] + times
    xs, t = np.meshgrid(np.arange(n), times)
    xyt_db = np.hstack((xs.flatten()[:, None], np.zeros(xs.shape).flatten()[:, None], t.flatten()[:, None]))
    
    # upper border points (x, n-1, t)
    xyt_ub = xyt_db + np.array([0, n-1, 0])
    
    # right border points (0, y, t)
    xyt_rb = np.hstack((np.zeros(xs.shape).flatten()[:, None], xs.flatten()[:, None], t.flatten()[:, None]))

    # left border points (n-1, y, t)
    xyt_lb = xyt_rb + np.array([n-1, 0, 0])

    # next step : collocation points
    lower_bound = np.array([0, 0, 0])
    upper_bound = np.array([n-1, n-1, max(times)])
    
    xyt_f = lower_bound + (upper_bound - lower_bound)*lhs(3, nb_colloc)
    
    # layers 
    layers = [3, 50, 50, 50, 50, 50, 50, 50, 50, 50, 1]

    plot_field_2D(initial_array, colorbar=True)
    
    # pass to the model
    model = PinnSH(xy0, xyt_db, xyt_ub, xyt_rb, xyt_lb, values, xyt_f, layers, epsilon, gamma)
    
    # traiiiining
    start = time.time()
    model.train(epochs=epochs, lr=lr)
    elapsed = time.time() - start
    print(f"Training time : {elapsed:.4f}")
    print(f"Error at training : {model.train_cost_history[-1]}")

    # Prediction on initial
    x = torch.tensor(xy0[:,0:1], dtype=dtype, device=device)
    y = torch.tensor(xy0[:,1:2], dtype=dtype, device=device)
    t = torch.tensor(xy0[:,2:3], dtype=dtype, device=device)
    
    # prediction
    u_predict = model.field_u(x,y,t).detach().cpu().numpy()
    u_predict = u_predict.reshape(n,n)

    error = np.linalg.norm(initial_array-u_predict,2) / np.linalg.norm(initial_array, 2)
    print(f"\n\n*********** Prediction error on initial : {error:.4f}")

    # plot on initial
    print("\nPlot on initial")
    plot_field_2D(u_predict, colorbar=True)

    # Generate prediction points
    xyt = make_test(n, max(times))
    xyt = make_test(n, 0)
    x = torch.tensor(xyt[:,0:1], dtype=dtype, device=device)
    y = torch.tensor(xyt[:,1:2], dtype=dtype, device=device)
    t = torch.tensor(xyt[:,2:3], dtype=dtype, device=device)
    print(x.shape, y.shape, t.shape)
    
    # prediction
    u_predict = model.field_u(x,y,t).detach().cpu().numpy()
    u_predict = u_predict.reshape(n,n)

    # field found by numerical solver
    final = np.load("/content/Data/field2.npy")
    
    # compute error
    error = np.linalg.norm(final-u_predict,2) / np.linalg.norm(final, 2)
    print(f"Prediction error : {error:.4f}")

    # plot the fields
    # plot_field_2D(final, colorbar=True) # , save_path="/content/Data/Figs"
    plot_field_2D(u_predict, colorbar=True)
```
